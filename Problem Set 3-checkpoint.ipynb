# Step 5. Discover the Male ratio per occupation and sort it from the most to the least


```python
# create a function to convert M and F into integer 1 and 0
def gender_to_numeric(x):
    if x ==  'M':
        return 1
    if x == 'F':
        return 0
```

```python
# apply the function to the gender column and create a new column "gender_n"
users['gender_n'] = users['gender'].apply(gender_to_numeric)
```

```python
male_ratio = users.groupby('occupation').gender_n.sum() /users.occupation.value_counts() * 100
male_occupations = male_ratio.sort_values(ascending = False)
male_ratio.sort_values(ascending = False)
male_occupations
```

```python
print(users.groupby(['occupation'])['age'].agg(['min', 'max']))
```

# Step 7. For each combination of occupation and sex, calculate the mean age

```python
print(users.groupby(['occupation', 'gender']).age.mean())
```

# Step 8. For each occupation present the percentage of women and men


```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})

occup_count = users.groupby(['occupation']).count()

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup /occup_count * 100
occup_gender.loc[:, 'gender']
```
